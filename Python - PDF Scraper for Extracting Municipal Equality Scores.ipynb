```python
# make a list of all of the urls

url = 'https://www.hrc.org/resources/mei-2023-see-your-cities-scores'

# Fetch the webpage content
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')

# Find the HTML elements containing city URLs
city_elements = soup.find_all('a', class_='block-link')

city_urls = [link['href'] for link in city_elements]

city_urls
```

```python
# define function to get pdf

def get_pdf(city):

    # url = city
    req = requests.get(city)
    
    # Find the position of the last '/'
    last_slash_index = url.rfind('/')
    
    # Extract the substring after the last '/'
    if last_slash_index != -1:  # Ensure '/' was found
        pdf_name = url[last_slash_index + 1:]
    else:
        pdf_name = url  # Handle case where '/' is not found (unlikely in a valid URL)

    # return pdf_name

    if req.status_code == 200:
     # Save the PDF content to a file
        with open(pdf_name, 'wb') as f:
            f.write(req.content)

     # Open the downloaded PDF file
    with open(pdf_name, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)

         # Loop through each page and extract text
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text = page.extract_text()

    return text

get_pdf('https://hrc-prod-requests.s3-us-west-2.amazonaws.com/MEI-2023-Assets/MEI-2023-Gainesville-Florida.pdf')
```
